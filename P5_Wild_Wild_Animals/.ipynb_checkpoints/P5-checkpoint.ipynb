```python


classes = ["Bighorn_Sheep", "Bobcat" ,"Coyote" ,"Gray_Fox","Javelina", "Mule_Deer", "Raptor","White_tailed_Deer"]

import tensorflow as tf
gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)

    except RuntimeError as e:
        print(e)
```

```python
def generate_k_fold_cross_validation_sets(X, T, n_folds, shuffle=True):
    if shuffle:
        # Randomly order X and T
        randorder = np.arange(X.shape[0])
        np.random.shuffle(randorder)
        X = X[randorder, :]
        T = T[randorder, :]

    # Partition X and T into folds
    n_samples = X.shape[0]
    n_per_fold = round(n_samples / n_folds)
    n_last_fold = n_samples - n_per_fold * (n_folds - 1)

    folds = []
    start = 0
    for foldi in range(n_folds-1):
        folds.append( (X[start:start + n_per_fold, :], T[start:start + n_per_fold, :]) )
        start += n_per_fold
    folds.append( (X[start:, :], T[start:, :]) )

    # Yield k(k-1) assignments of Xtrain, Train, Xvalidate, Tvalidate, Xtest, Ttest

    for validation_i in range(n_folds):
        for test_i in range(n_folds):
            if test_i == validation_i:
                continue

            train_i = np.setdiff1d(range(n_folds), [validation_i, test_i])

            Xvalidate, Tvalidate = folds[validation_i]
            Xtest, Ttest = folds[test_i]
            if len(train_i) > 1:
                Xtrain = np.vstack([folds[i][0] for i in train_i])
                Ttrain = np.vstack([folds[i][1] for i in train_i])
            else:
                Xtrain, Ttrain = folds[train_i[0]]

            yield Xtrain, Ttrain, Xvalidate, Tvalidate, Xtest, Ttest
```

```python
def make_CNN_Model(x_train,hidden_layers,af,num_class):
    model = models.Sequential()
    model.add(layers.Conv2D(hidden_layers[0], (3, 3), activation=af, input_shape=x_train[0].shape))
    for i in range(1,len(hidden_layers)):
        model.add(layers.MaxPooling2D(2,2))
        model.add(layers.Conv2D(hidden_layers[i], (3, 3), activation=af, input_shape=x_train[0].shape))
    model.add(layers.Flatten())
    model.add(layers.Dense(128,activation=af))
    model.add(layers.Dense(num_class))
    model.compile(optimizer='adam', loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
    return model
```

```python
def load_data(f):
    pick = open(f, 'rb')
    data = pickle.load(pick)
    pick.close()
    features = []
    labels = []
    for img, label in data:
        features.append(np.array(img))
        labels.append(np.array(label))
    features = np.asarray(features).astype(np.float32)
    feature = features / 255.
    labels = np.array(labels)
    return feature,labels
```

```python
def run_CNN(X,T,arch,epochs, folds=3,af='relu'):
    n_classes = len(np.unique(T))
    trainA = []
    testA  = []
    valA   = []
    Xt=None
    Tt=None
    m=None
    print("dsada")
    for Xtrain, Ttrain, Xval, Tval, Xtest, Ttest in generate_k_fold_cross_validation_sets(X, T, folds, shuffle=True):
        print("Here")
        model = make_CNN_Model(X,arch,af,n_classes)
        if(folds ==1):
            history = model.fit(Xtrain,Ttrain, epochs=epochs, validation_split=0.2)
            print("here")
        else:
            history = model.fit(Xtrain,Ttrain, epochs=epochs, validation_data=(Xval, Tval))
        eva = model.evaluate(Xtest, Ttest)
        acc = history.history['accuracy']
        val_acc = history.history['val_accuracy']
        trainA.append(acc[-1])
        testA.append(eva[1])
        valA.append(val_acc[-1])
        Xt=Xtest
        Tt=Ttest
        m=model
    return np.mean(trainA), np.mean(testA), np.mean(valA), Xt, Tt,m
```

```python
def run_pretrained(X,T,folds,epochs,network):
    n_classes = len(np.unique(T))
    trainA = []
    testA  = []
    valA   = []
    Xt=None
    Tt=None
    m=None
    for Xtrain, Ttrain, Xval, Tval, Xtest, Ttest in generate_k_fold_cross_validation_sets(X, T, folds, shuffle=True):
            base_model = network(weights = 'imagenet', include_top = False, input_shape=X[0].shape)
            x = base_model.output
            x = layers.GlobalAveragePooling2D()(x)
            x = layers.Dense(1024, activation='relu')(x)
            predictions = layers.Dense(len(np.unique(T)), activation='softmax')(x)
            model = Model(inputs=base_model.input, outputs=predictions)
            for layer in base_model.layers:
                          layer.trainable = False
            model.compile(optimizer='rmsprop', loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
                     metrics=['accuracy'])
                          
                          
            history = model.fit(Xtrain,Ttrain, epochs=epochs, validation_data=(Xval, Tval))
            for layer in base_model.layers:
                          layer.trainable = True
            model.compile(optimizer=SGD(lr=0.0001, momentum=0.9), loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
                     metrics=['accuracy'])
            history = model.fit(Xtrain,Ttrain, epochs=epochs, validation_data=(Xval, Tval))
            eva = model.evaluate(Xtest, Ttest)
            acc = history.history['accuracy']
            val_acc = history.history['val_accuracy']
            trainA.append(acc[-1])
            testA.append(eva[1])
            valA.append(val_acc[-1])
            Xt=Xtest
            Tt=Ttest
            m=model
    return np.mean(trainA), np.mean(testA), np.mean(valA), Xt, Tt ,m
```

```python
def get_label(file):
    for c in classes:
        if c in file:
            return classes.index(c)
```

```python
def remove_col(dt):
    s = copy.deepcopy(dt)
    for x in s:
        del x[0:-1]
    return s
```

```python
def run_CNN_experiment(X,T,archs,epochs,folds=3,af='relu'):
    dt=[]
    confs = []
    for arch in archs:
        print(f"On arch: {arch}")
        train_acc, test_acc,val_acc,Xtest,Ttest,model = run_CNN(X,T,arch,epochs)
        dt.append([arch,epochs, train_acc,val_acc,test_acc])
        predictions = tf.argmax(tf.nn.softmax(model.predict(Xtest.astype('float32'))), axis=1)
        confusion = tf.math.confusion_matrix(labels=Ttest, predictions=predictions,num_classes=len(np.unique(Ttest)))
        confusion = pd.DataFrame(confusion.numpy(), index=classes, columns=classes)
        confusion = confusion.div(confusion.sum(axis=1),axis=0)*100
        confusion = confusion.style.background_gradient(cmap='Blues').format("{:.1f} %")
        confs.append(confusion)
    val = np.argmax(remove_col(dt))
    return pd.DataFrame(dt,columns=["architecture", "Epochs","Train Accuracy", "Validation Accuracy","Test Accuracy"]),confs[val]
```

```python
def run_PreTrained_experiment(X, T, folds, networks, epochs):
    dt=[]
    confs = []
    for network in networks:
        print(f"On network: {network}")
        train_acc, test_acc,val_acc,Xtest,Ttest,model = run_pretrained(X,T,folds,epochs,network)
        dt.append([network,epochs, train_acc,val_acc,test_acc])
        predictions = tf.argmax(tf.nn.softmax(model.predict(Xtest.astype('float32'))), axis=1)
        confusion = tf.math.confusion_matrix(labels=Ttest, predictions=predictions,num_classes=len(np.unique(Ttest)))
        confusion = pd.DataFrame(confusion.numpy(), index=classes, columns=classes)
        confusion = confusion.div(confusion.sum(axis=1),axis=0)*100
        confusion = confusion.style.background_gradient(cmap='Blues').format("{:.1f} %")
        confs.append(confusion)
    val = np.argmax(remove_col(dt))
    return pd.DataFrame(dt,columns=["architecture", "Epochs","Train Accuracy", "Validation Accuracy","Test Accuracy"]),confs[val]
```

```python
def run_CNN_ensemble(X,T,arch,epochs,af='relu'):
    n_classes = len(np.unique(T))

    x_train, x_test, y_train,y_test = train_test_split(X,T,test_size=0.1,shuffle=True)
    model = make_CNN_Model(X,arch,af,n_classes)
    history = model.fit(x_train,y_train, epochs=epochs, validation_split=0.2)
    eva = model.evaluate(x_test, y_test)
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    return acc[-1], eva[1], val_acc[-1], x_test,y_test,model
```

```python
def chunks(lst, n):
    """Yield successive n-sized chunks from lst."""
    for i in range(0, len(lst), n):
        yield lst[i:i + n]
```

```python
def Ensemble_list_of_models(X,T,archs,epochs):
    #print(f"Len X: {len(X)}, Len T: {len(T)}")
    T= T.reshape((-1,1))
    randorder = np.arange(X.shape[0])
    np.random.shuffle(randorder)
    X = X[randorder, :]
    T = T[randorder, :]
    num_voters = len(archs)
    constituents = []
    x_ts = [t for t in chunks(X,len(X)//num_voters)]
    t_ts = [t for t in chunks(T,len(T)//num_voters)]
    for arch,x_t,t_t in zip(archs,x_ts,t_ts):
        train_acc, test_acc,val_acc,Xt,Tt,model = run_CNN_ensemble(x_t,t_t,arch,epochs=epochs)
        constituents.append((model,test_acc))
    return constituents
        
```

```python
def Ensemble_Eval(X,T,X_test,T_test, archs,epochs):
    correct = 0 
    constituents = Ensemble_list_of_models(X,T,archs,epochs)
    preditions = []
    acc = []
    for m,T_a in constituents:
        p = tf.argmax(tf.nn.softmax(m.predict(X_test)), axis=1)
        preditions.append(p.numpy())
    #print(preditions)
    pred = np.asarray(preditions).transpose()
    #print(pred)
    for i in range(len(pred)):
        val = mode(pred[i])
        if val == T_test[i]:
            correct +=1
    return correct/len(T_test)
    
```
